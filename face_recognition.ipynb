# Teste

```python
# Load an image with an unknown face
unknown_image = face_recognition.load_image_file("trump_marcelo2.jpg")

# Find all the faces and face encodings in the unknown image
face_locations = face_recognition.face_locations(unknown_image)
face_encodings = face_recognition.face_encodings(unknown_image, face_locations)

# Convert the image to a PIL-format image so that we can draw on top of it with the Pillow library
# See http://pillow.readthedocs.io/ for more about PIL/Pillow
pil_image = Image.fromarray(unknown_image)
# Create a Pillow ImageDraw Draw instance to draw with
draw = ImageDraw.Draw(pil_image)


flag = 0
autorizado = 0

# Loop through each face found in the unknown image
for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
    
    flag = 0
    
    # See if the face is a match for the known face(s)
    matches = face_recognition.compare_faces(known_face_encodings, face_encoding)

    name = "Desconhecido" 

    # Or instead, use the known face with the smallest distance to the new face
    face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
    best_match_index = np.argmin(face_distances)
    if matches[best_match_index]:
        name = known_face_names[best_match_index]
        
        if dict.get(name) == 1:
            flag = 1
        elif dict.get(name) == 2: 
            flag = 2
        elif dict.get(name) == 3:
            flag = 3
        elif dict.get(name) == 4:
            flag = 4
        elif dict.get(name) == 5:
            flag = 5
        elif dict.get(name) == 6:
            flag = 6
    
    if flag == 0:    #DESCONHECIDO
        # Draw a box around the face using the Pillow module
        draw.rectangle(((left, top), (right, bottom)), outline=(220, 200, 0))
    
        # Draw a label with a name below the face
        text_width, text_height = draw.textsize(name)
        draw.rectangle(((left, bottom - text_height - 10), (right, bottom)), fill=(220, 200, 0), outline=(220, 200, 0))
        draw.text((left + 6, bottom - text_height - 5), name, fill=(255, 255, 255, 255))
        
    elif flag == 1:  #NÃO TEM PROJETO
        # Draw a box around the face using the Pillow module
        draw.rectangle(((left, top), (right, bottom)), outline=(0, 0, 255))
    
        # Draw a label with a name below the face
        text_width, text_height = draw.textsize(name)
        draw.rectangle(((left, bottom - text_height - 10), (right, bottom)), fill=(0, 0, 255), outline=(0, 0, 255))
        draw.text((left + 6, bottom - text_height - 5), name + "\nEm espera!", fill=(255, 255, 255, 255))
        flag = 0
    
    elif flag == 2:  #BOSS
        # Draw a box around the face using the Pillow module
        draw.rectangle(((left, top), (right, bottom)), outline=(0, 200, 0))
    
        # Draw a label with a name below the face
        text_width, text_height = draw.textsize(name)
        draw.rectangle(((left, bottom - text_height - 10), (right, bottom)), fill=(0, 200, 0), outline=(0, 200, 0))
        draw.text((left + 6, bottom - text_height - 5), name + "\nBoss", fill=(255, 255, 255, 255))
        flag = 0
    
    elif flag == 3: #JÁ ESTÁ COLOCADO
        # Draw a box around the face using the Pillow module
        draw.rectangle(((left, top), (right, bottom)), outline=(0, 70, 0))
    
        # Draw a label with a name below the face
        text_width, text_height = draw.textsize(name)
        draw.rectangle(((left, bottom - text_height - 10), (right, bottom)), fill=(0, 70, 0), outline=(0, 70, 0))
        draw.text((left + 6, bottom - text_height - 5), name + "\nColocado", fill=(255, 255, 255, 255))
        flag = 0
        
    elif flag == 4: #DESISTIU 
        draw.rectangle(((left, top), (right, bottom)), outline=(255, 0, 0))
        text_width, text_height = draw.textsize(name)
        draw.rectangle(((left, bottom - text_height - 10), (right, bottom)), fill=(255, 0, 0), outline=(255, 0, 0))
        draw.text((left + 6, bottom - text_height - 5), name + "\nDesistiu", fill=(255, 255, 255, 255))
        flag = 0
        
    elif flag == 5: #Perigo público 
        draw.rectangle(((left, top), (right, bottom)), outline=(255, 0, 0))
        text_width, text_height = draw.textsize(name)
        draw.rectangle(((left, bottom - text_height - 10), (right, bottom)), fill=(255, 0, 0), outline=(255, 0, 0))
        draw.text((left + 6, bottom - text_height - 5), name + "\nPerigo\nPúblico", fill=(255, 255, 255, 255))
        flag = 0
        
    elif flag == 6: #Presidente da republica 
        draw.rectangle(((left, top), (right, bottom)), outline=(0, 0, 240))
        text_width, text_height = draw.textsize(name)
        draw.rectangle(((left, bottom - text_height - 10), (right, bottom)), fill=(0, 0, 240), outline=(0, 0, 240))
        draw.text((left + 6, bottom - text_height - 5), name + "\nPresidente", fill=(255, 255, 255, 255))
        flag = 0
        autorizado = 1
        
# Remove the drawing library from memory as per the Pillow docs
del draw

# Display the resulting image
#display(pil_image)

if autorizado == 1:
    print("Presidente da república identificado.\nAutorização concedida.")
    display(pil_image)
    
```
